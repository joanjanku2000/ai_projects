```python
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

def root_relative_squared_error(actual, predicted):
    mse = mean_squared_error(actual, predicted)
    mean_y = np.mean(actual)
    tss = np.sum((actual - mean_y) ** 2)
    rrse = np.sqrt(mse / tss)
    return rrse
def root_absolute_error(actual,predicted):
  diff = np.abs(actual - predicted)
  mean_diff = np.abs(actual - np.mean(actual))
  rae = np.sum(diff) / np.sum(mean_diff)
  return rae.mean()
```

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, cohen_kappa_score, mean_absolute_error, mean_squared_error, confusion_matrix
from sklearn.model_selection import train_test_split

data = pd.read_csv('/content/spambase_binary.csv', header=None)

# Split the data into features (X) and labels (y)
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
# Train the Naive Bayes classifier using 10-fold cross-validation
gnb = GaussianNB()

# Iterations of 10-fold cross-validation
iterations = [10, 50, 100]
gnb.fit(X_train, y_train)
for i in iterations:
    scores = cross_val_score(gnb, X, y, cv=i)

    # Calculate the accuracy
    acc = np.mean(scores)

    y_pred = gnb.predict(X_test);
    # Calculate the number of correctly and incorrectly classified instances
    correct = (y_pred == y_test).sum()
    incorrect = (y_pred != y_test).sum()

    # Calculate the Cohen Kappa Score
    kappa = cohen_kappa_score(y_test, y_pred)

    # Calculate the Mean Absolute Error
    mae = mean_absolute_error(y_test, y_pred)

    # Calculate the Root Mean Squared Error
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    # Calculate the Relative Absolute Error
    rae = root_absolute_error(y_test,y_pred)
    # Calculate the Root Relative Squared Error
    rrse = root_relative_squared_error(y_test,y_pred)

    # Print the results for each iteration
    print("Iteration: ", i)
    print("Accuracy: ", acc)
    print("Correctly classified instances: ", correct)
    print("Incorrectly classified instances: ", incorrect)
    print("Cohen Kappa Score: ", kappa)
    print("Mean Absolute Error: ", mae)
    print("Root Mean Squared Error: ", rmse)
    print("Relative Absolute Error: ", rae)
    print("Root Relative Squared Error: ", rrse)
    
print("========= PERCEPTRON =========")
from sklearn.linear_model import Perceptron
perceptron = Perceptron()
perceptron.fit(X_train,y_train)
for i in iterations:
    scores = cross_val_score(perceptron, X, y, cv=i)

    # Calculate the accuracy
    acc = np.mean(scores)

    y_pred = perceptron.predict(X_test);
    # Calculate the number of correctly and incorrectly classified instances
    correct = (y_pred == y_test).Ssum()
    incorrect = (y_pred != y_test).sum()

    # Calculate the Cohen Kappa Score
    kappa = cohen_kappa_score(y_test, y_pred)

    # Calculate the Mean Absolute Error
    mae = mean_absolute_error(y_test, y_pred)

    # Calculate the Root Mean Squared Error
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    # Calculate the Relative Absolute Error
    rae = root_absolute_error(y_test,y_pred)
    # Calculate the Root Relative Squared Error
    rrse = root_relative_squared_error(y_test,y_pred)

    # Print the results for each iteration
    print("Iteration: ", i)
    print("Accuracy: ", acc)
    print("Correctly classified instances: ", correct)
    print("Incorrectly classified instances: ", incorrect)
    print("Cohen Kappa Score: ", kappa)
    print("Mean Absolute Error: ", mae)
    print("Root Mean Squared Error: ", rmse)
    print("Relative Absolute Error: ", rae)
    print("Root Relative Squared Error: ", rrse)


print("========= SVC =========")

from sklearn.svm import SVC
svm = SVC(kernel='linear')
svm.fit(X_train,y_train)
for i in iterations:
    scores = cross_val_score(svm, X, y, cv=i)

    # Calculate the accuracy
    acc = np.mean(scores)

    y_pred = svm.predict(X_test);
    # Calculate the number of correctly and incorrectly classified instances
    correct = (y_pred == y_test).sum()
    incorrect = (y_pred != y_test).sum()

    # Calculate the Cohen Kappa Score
    kappa = cohen_kappa_score(y_test, y_pred)

    # Calculate the Mean Absolute Error
    mae = mean_absolute_error(y_test, y_pred)

    # Calculate the Root Mean Squared Error
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    # Calculate the Relative Absolute Error
    rae = root_absolute_error(y_test,y_pred)
    # Calculate the Root Relative Squared Error
    rrse = root_relative_squared_error(y_test,y_pred)

    # Print the results for each iteration
    print("Iteration: ", i)
    print("Accuracy: ", acc)
    print("Correctly classified instances: ", correct)
    print("Incorrectly classified instances: ", incorrect)
    print("Cohen Kappa Score: ", kappa)
    print("Mean Absolute Error: ", mae)
    print("Root Mean Squared Error: ", rmse)
    print("Relative Absolute Error: ", rae)
    print("Root Relative Squared Error: ", rrse)


```
